```python
import os
import re
from PIL import Image
```

```python
root_path = "E:/data/mvtec"
```

```python
source_dir = os.path.join(root_path, 'scenes')
train_path = os.path.join(root_path, 'train')
print(source_dir)
print(train_path)
```

```python
entries = os.listdir(source_dir)
print(len(entries))
print(entries[:10])
```

```python
for entry in entries:
    file_path = os.path.join(source_dir, entry)
    for file in os.listdir(file_path):
        if os.path.isfile(os.path.join(file_path, file)):
            if file.lower().find('_l.tif') > 0:
                img = Image.open(os.path.join(file_path, file))
                new_file = file.replace(".tif", ".jpg")
                new_file = entry + '-' + new_file
                img.save(os.path.join(train_path, new_file))
               
```

```python
labels_path = os.path.join(root_path, 'scene_lists')
labels_path
```

```python
entries = os.listdir(labels_path)
print(len(entries))
print(entries[:10])
```

```python
labels = {}
classes = set()
for entry in entries:
    if os.path.isfile(os.path.join(labels_path, entry)):
        class_value = entry.replace(".txt", "")
        class_value = class_value.replace("scene_list_", "")
        classes.add(class_value)
        print(class_value)
        file = open(os.path.join(labels_path, entry), 'r')
        for line in file:
            try:
                key = int(line)
                if key in labels:
                    labels[key].append(class_value)
                else:
                    labels[key] = []
                    labels[key].append(class_value)
            except ValueError:
                continue
print(len(classes))
print(len(labels))
```

```python
import json
```

```python
with open(os.path.join(root_path, "classes.json"), 'w') as output:
    json.dump(labels, output)
```

```python
import shutil
```

```python
entries = os.listdir(train_path)
for entry in entries:
    if os.path.isfile(os.path.join(train_path, entry)):
        key = entry[6:10]
        try:
            key = int(key)
            class_list = labels[key]
            label = class_list[0]
            curr_path = os.path.join(train_path, entry)
            dest_path = os.path.join(train_path, label)
            dest_path = os.path.join(dest_path, entry)
            shutil.move(curr_path, dest_path)
        except ValueError:
            continue
    
```

```python
test_path = os.path.join(root_path, 'test')
```

```python
from random import sample
```

```python
entries = os.listdir(train_path)
for entry in entries:
    if os.path.isdir(os.path.join(train_path, entry)):
        os.mkdir(os.path.join(test_path, entry))
        curr_dir = os.path.join(train_path, entry)
        dest_dir = os.path.join(test_path, entry)
        files = os.listdir(curr_dir)
        files = [f for f in files if os.path.isfile(os.path.join(curr_dir, f))]
        selected = sample(files, k=3)
        for file in selected:
            curr_path = os.path.join(curr_dir, file)
            dest_path = os.path.join(dest_dir, file)
            shutil.move(curr_path, dest_path)
```
